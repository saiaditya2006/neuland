```python
import pandas as pd
from fuzzywuzzy import fuzz
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
import re

# Load the Excel file
df = pd.read_excel("neudata.xlsx")

# Step 1: Clean and standardize for fuzzy matching
df['HOLDER'] = df['HOLDER'].astype(str)

# Step 2: Normalize names: lowercase, strip, and remove common suffixes
def normalize_name(name):
    name = name.lower().strip()
    name = re.sub(r'\b(inc|inc\.|ltd|ltd\.|llc|corp|corporation|co\.|company|plc|gmbh|s\.a\.)\b', '', name)
    name = re.sub(r'[^\w\s]', '', name)  # remove punctuation
    return name.strip()

df['HOLDER_clean'] = df['HOLDER'].apply(normalize_name)
df['HOLDER_standardized'] = df['HOLDER_clean']

# Step 3: Fuzzy compare all pairs with a threshold
threshold = 70
for i, val_i in enumerate(df['HOLDER_clean']):
    for j, val_j in enumerate(df['HOLDER_clean']):
        if i != j:
            score = fuzz.token_sort_ratio(val_i, val_j)
            if score >= threshold:
                print(f"Row {i} ('{val_i}') matches Row {j} ('{val_j}') with score: {score}")
```

```python
#Close to final --1
import pandas as pd
from fuzzywuzzy import fuzz

# Load the Excel file
df = pd.read_excel("neudata.xlsx")

# Step 1: Clean and standardize for fuzzy matching
df['HOLDER'] = df['HOLDER'].astype(str)
df['HOLDER_clean'] = df['HOLDER'].str.lower().str.strip()
df['HOLDER_standardized'] = df['HOLDER_clean']

# Initialize an empty list to hold the clusters of similar values
clusters = []

threshold = 80

# Step 2: Iterate through the rows and perform fuzzy matching
for i, val_i in enumerate(df['HOLDER_clean']):
    matched = False
    for cluster in clusters:
        for val_j in cluster:
            score = fuzz.token_sort_ratio(val_i, val_j)
            if score >= threshold:
                # Add val_i to the current cluster and mark as matched
                cluster.append(val_i)
                matched = True
                break
        if matched:
            break
    
    if not matched:
        # If no match was found, create a new cluster
        clusters.append([val_i])

# Step 3: Print the clusters to verify
for idx, cluster in enumerate(clusters):
    print(f"Cluster {idx + 1}: {cluster}")

# Now you have a list of clusters where each cluster contains similar values
```

```python
#Close to final
import pandas as pd
from fuzzywuzzy import fuzz
import re

# Load the Excel file
df = pd.read_excel("neudata.xlsx")

# Define stopwords to clean
company_stopwords = [
    r'\bco\b', r'\bltd\b', r'\binc\b', r'\bproducts\b',
    r'\bcorporation\b', r'\bcorp\b', r'\bprivate\b', r'\blimited\b',
    r'\bplc\b', r'\bpvt\b', r'\bgroup\b', r'\bcompany\b', r'\bchemical\b', r'\bchemicals'
]

# Function to remove stopwords and clean names
def clean_company_name(name):
    name = name.lower()
    name = re.sub('|'.join(company_stopwords), '', name)
    name = re.sub(r'[^a-z0-9\s]', '', name)
    name = re.sub(r'\s+', ' ', name).strip()
    return name

# Clean and standardize for fuzzy matching
df['HOLDER'] = df['HOLDER'].astype(str)
df['HOLDER_clean'] = df['HOLDER'].apply(clean_company_name)

# Cluster similar names
clusters = []
threshold = 80

for i, val_i in enumerate(df['HOLDER_clean']):
    matched = False
    for cluster in clusters:
        for val_j in cluster:
            score = fuzz.token_sort_ratio(val_i, val_j)
            if score >= threshold:
                cluster.append(val_i)
                matched = True
                break
        if matched:
            break
    if not matched:
        clusters.append([val_i])

# Print clusters
for idx, cluster in enumerate(clusters):
    print(f"Cluster {idx + 1}: {cluster}")
```
